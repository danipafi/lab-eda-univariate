### Part 1: Understanding Product Categories

**Business Question**: What are the most popular product categories on Amazon UK, and how do they compare in terms of listing frequency?

1. **Frequency Tables**:
    - Generate a frequency table for the product `category`.
    - Which are the top 5 most listed product categories?

2. **Visualizations**:
    - Display the distribution of products across different categories using a bar chart. *If you face problems understanding the chart, do it for a subset of top categories.*
    - For a subset of top categories, visualize their proportions using a pie chart. Does any category dominate the listings?

```python
# Frequency table for 'category'
frequency_table = df['category'].value_counts()

# Calculating the proportion of each unique value
proportion_table = df['category'].value_counts(normalize=True)

frequency_table, proportion_table
```

```python
# Creating a crosstab table for the 'category' column, counting occurrences for each unique value
my_table = pd.crosstab(index = df["category"],  # Make a crosstab
                              columns="count")      # Name the count column
my_table = my_table.sort_values(by="count", ascending=False)
top_5 = my_table.head(5)
top_5
```

```python
# Plotting a bar chart using the values from the frequency table, with colors sourced from the "Set3" palette
sns.barplot(x=frequency_table.index, y=frequency_table.values, palette="Set3");
```

```python
# Plotting a bar chart using the 'count' values from 'my_table', with colors sourced from the "Set3" palette
sns.barplot(x=top_5.index, y=top_5["count"], palette="Set3");
```

```python
top_5.reset_index(inplace=True)
```

```python
top_5
```

```python
# Criar um pie chart com os dados corretos
plt.figure(figsize=(8, 8))
plt.pie(
    top_5['count'], 
    labels=top_5['category'], 
    autopct='%1.1f%%', 
    startangle=90, 
    colors=sns.color_palette("Set3")
)
```

```python
# Plotting a pie chart of the 'count' column with percentage labels
# starting at angle 90, and using colors from the "Set3" Seaborn palette
plt.figure(figsize=(8, 8))
plt.pie(
    top_5['count'], 
    labels=top_5['category'], 
    autopct='%1.1f%%', 
    startangle=90, 
    colors=sns.color_palette("Set3")
)
```

The Sports & Outdoors category dominates Amazon UK with 836,265 listings, far surpassing other categories. The remaining top 4 categories (Beauty, Handmade Clothing, Bath & Body, Birthday Gifts) have similar listing counts, around 19K each. This highlights a massive gap, suggesting Sports & Outdoors has significantly higher demand or seller interest.

### Part 2: Delving into Product Pricing

**Business Question**: How are products priced on Amazon UK, and are there specific price points or ranges that are more common?

1. **Measures of Centrality**:
    - Calculate the mean, median, and mode for the `price` of products.
    - What's the average price point of products listed? How does this compare with the most common price point (mode)?

2. **Measures of Dispersion**:
    - Determine the variance, standard deviation, range, and interquartile range for product `price`.
    - How varied are the product prices? Are there any indicators of a significant spread in prices?

3. **Visualizations**:
    - Is there a specific price range where most products fall? Plot a histogram to visualize the distribution of product prices. *If its hard to read these diagrams, think why this is, and explain how it could be solved.*.
    - Are there products that are priced significantly higher than the rest? Use a box plot to showcase the spread and potential outliers in product pricing. 


```python
# measures of centrality

mean_price = df['price'].mean()
median_price = df['price'].median()
mode_price = df['price'].mode()[0]

mean_price, median_price, mode_price
```

The **average price** of listed products is **£89.24**, while the **most common price** (mode) is significantly lower at **£9.99**. This suggests that although some high-priced products raise the mean, the majority of listings are at much lower price points. The **median price (£19.09)** further confirms that most products are relatively inexpensive, as half of the listings are priced below this value.

```python
# measures of dispersion

variance_price = df['price'].var()
std_dev_price = df['price'].std()
min_price = df['price'].min()
max_price = df['price'].max()
range_price = max_price - min_price
interquartile = df['price'].quantile(0.75) - df['price'].quantile(0.25)

variance_price, std_dev_price, min_price, max_price, range_price, interquartile
```

Product prices are highly varied, with extreme outliers inflating the range (£0 to £100,000) and variance. Despite this, most prices fall within a much smaller range (IQR = £36), suggesting that the majority of listings are relatively affordable.

```python
sns.histplot(df['price'], kde=True, bins=3, color="salmon")
```

```python
# Limiting the price values to not show extreme outliers
price_limit = df['price'].quantile(0.99)  # Limit up to the 99th percentile
df_filtered = df[df['price'] <= price_limit]

# Adjusting the histogram visualization with bins and KDE
plt.figure(figsize=(12, 6))

# Creating the histogram with adjusted bins
sns.histplot(df_filtered['price'], bins=50, kde=False, color="skyblue")

# Title and labels
plt.title('Product Price Distribution (Without Extreme Outliers)', fontsize=14)
plt.xlabel('Price', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Displaying the plot
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```

To better visualize the price distribution for most products, we limited the product prices by filtering out the top 1% of extreme values (outliers) using the 99th percentile. Then we created a histogram of the filtered data, using 50 bins, which shows the frequency distribution of the product prices.

```python
# Plotting a boxplot for the 'price' column with a light blue color
sns.boxplot(data = df['price'], color="lightblue")
```

```python
# Limiting price values to exclude extreme outliers
price_limit = df['price'].quantile(0.99)  # Limit to the 99th percentile
df_filtered = df[df['price'] <= price_limit]

# Creating the boxplot with filtered prices
sns.boxplot(data = df_filtered['price'], color="skyblue")
```

We filtered out extreme outliers and created a boxplot to better visualize the distribution of product prices.

### Part 3: Unpacking Product Ratings

**Business Question**: How do customers rate products on Amazon UK, and are there any patterns or tendencies in the ratings?

1. **Measures of Centrality**:
    - Calculate the mean, median, and mode for the `rating` of products.
    - How do customers generally rate products? Is there a common trend?

2. **Measures of Dispersion**:
    - Determine the variance, standard deviation, and interquartile range for product `rating`.
    - Are the ratings consistent, or is there a wide variation in customer feedback?

3. **Shape of the Distribution**:
    - Calculate the skewness and kurtosis for the `rating` column. 
    - Are the ratings normally distributed, or do they lean towards higher or lower values?

4. **Visualizations**:
    - Plot a histogram to visualize the distribution of product ratings. Is there a specific rating that is more common?

```python
df.columns
```

```python
# measures of centrality

mean_rating = df['stars'].mean()
median_rating = df['stars'].median()
mode_rating = df['stars'].mode()[0]

mean_rating, median_rating, mode_rating
```

The trend we observe here is that most products have not been rated, resulting in a large number of "0" values in the dataset. When we calculate the mean and median ratings, these 0 values skew the results.

```python
# Filter out products with no ratings (0 ratings)
df_rated = df[df['stars'] > 0]

# Recalculate mean, median, and mode for rated products
mean_rating = df_rated['stars'].mean()
median_rating = df_rated['stars'].median()
mode_rating = df_rated['stars'].mode()[0]

mean_rating, median_rating, mode_rating
```

For the products that have been rated, customers generally give positive feedback, with a high concentration of ratings around 4.5. The trend shows that customers tend to rate products highly, and the ratings are consistently above 4, indicating a general satisfaction with the products.

```python
# measures of dispersion

variance_rating = df['stars'].var()
std_dev_rating = df['stars'].std()
min_rating = df['stars'].min()
max_rating = df['stars'].max()
range_rating = max_rating - min_rating
interquartile = df['stars'].quantile(0.75) - df['stars'].quantile(0.25)

variance_rating, std_dev_rating, min_rating, max_rating, range_rating, interquartile
```

```python
# Filter out products with no ratings (0 ratings)
df_rated = df[df['stars'] > 0]

# Recalculate mean, median, and mode for rated products
variance_rating = df_rated['stars'].var()
std_dev_rating = df_rated['stars'].std()
min_rating = df_rated['stars'].min()
max_rating = df_rated['stars'].max()
range_rating = max_rating - min_rating
interquartile = df_rated['stars'].quantile(0.75) - df_rated['stars'].quantile(0.25)

variance_rating, std_dev_rating, min_rating, max_rating, range_rating, interquartile
```

```python
from scipy.stats import skew, kurtosis

# Calculate skewness and kurtosis
skewness = skew(df['stars'])
kurt = kurtosis(df['stars'])

skewness, kurt
```

Skewness ≈ 0: The distribution of the ratings is approximately symmetrical, indicating no significant skew towards higher or lower values.

Kurtosis < 0: The distribution is flatter than normal, suggesting most ratings are close to the mean, with fewer extreme values.

```python
# Filter to exclude ratings with value 0
df_rated = df[df['stars'] > 0]

# Plot histogram of filtered ratings with 5 bins
sns.histplot(df_rated['stars'], bins=5, kde=False)

# Add labels and title
plt.xlabel('Product Ratings')
plt.ylabel('Frequency')
plt.title('Distribution of Product Ratings (Excluding 0)')

# Show the plot
plt.show()
```
